```python
import cv2
import tensorflow as tf

CATEGORIES = ['Fire_Alert', 'Normal']

def prepare(frame):
    IMG_SIZE = 60
    img_array = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

model = tf.keras.models.load_model("/home/navid/deep_learning/deepLearning/fire_detections/fireDetection_64x3-CNN.model")
```

```python
# Load video file
cap = cv2.VideoCapture('/home/navid/deep_learning/deepLearning/fire_detections/videos/forest2.mp4')  # Replace with your video file path
import time
# Get the video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
codec = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output_video.avi', codec, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Apply the model to the frame
    prediction = model.predict([prepare(frame)])
    
    # Get the predicted label
    label = CATEGORIES[int(prediction[0][0])]
    print(label)
    # If fire is detected, draw bounding box
    if label == 'Fire_Alert':
        cv2.rectangle(frame, (0, 0), (frame_width, frame_height), (0, 0, 255), 2)  # Draw a red bounding box

    # Add the label to the frame
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Write the frame with the label to the output video
    out.write(frame)

    # Display the frame
    cv2.imshow('Frame', frame)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
